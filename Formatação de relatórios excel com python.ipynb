# Criação de arquivo de excel

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font, Alignment, PatternFill
from openpyxl.utils import get_column_letter

# Exemplo de DataFrame
df = pd.DataFrame({
    "Nome": ["Coca-Cola", "Água Mineral", "Suco de Laranja"],
    "Categoria": ["Refrigerante", "Água", "Suco"],
    "Preço": [7.5, 2.0, 6.0]
})

arquivo = "relatorio_formatado.xlsx"

# 1. Salvar normalmente com pandas
df.to_excel(arquivo, index=False, sheet_name="Dados")

# 2. Abrir com openpyxl para formatar
wb = load_workbook(arquivo)
ws = wb["Dados"]

# --- CABEÇALHO ---
header_fill = PatternFill("solid", fgColor="D9D9D9")
header_font = Font(bold=True)
header_alignment = Alignment(horizontal="center", vertical="center")

for cell in ws[1]:
    cell.fill = header_fill
    cell.font = header_font
    cell.alignment = header_alignment

# --- FILTRO ---
ws.auto_filter.ref = ws.dimensions

# --- CONGELAR PRIMEIRA LINHA ---
ws.freeze_panes = "A2"

# --- AJUSTAR LARGURA DAS COLUNAS ---
for col in ws.columns:
    max_length = 0
    col_letter = get_column_letter(col[0].column)

    for cell in col:
        if cell.value:
            max_length = max(max_length, len(str(cell.value)))

    ws.column_dimensions[col_letter].width = max_length + 2

# 3. Salvar
wb.save(arquivo)

print("Excel gerado e formatado com sucesso!")
```

# Criando formulas em arquivos excel

```python
import pandas as pd
from openpyxl import load_workbook

df = pd.DataFrame({
    "Produto": ["Coca", "Água", "Suco"],
    "Quantidade": [2, 5, 3],
    "Preço": [7.5, 2.0, 6.0]
})

arquivo = "relatorio_com_formula.xlsx"

# Salva o DataFrame
df.to_excel(arquivo, index=False, sheet_name="Dados")

# Abre com openpyxl
wb = load_workbook(arquivo)
ws = wb["Dados"]

# Insere fórmula na coluna Total
ws["D1"] = "Total"

for row in range(2, ws.max_row + 1):
    ws[f"D{row}"] = f"=B{row}*C{row}"

wb.save(arquivo)

print("Arquivo gerado com fórmulas!")
```

# Formatando como tabela

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.worksheet.table import Table, TableStyleInfo

df = pd.DataFrame({
    "Produto": ["Coca", "Água", "Suco"],
    "Quantidade": [2, 5, 3],
    "Preço": [7.5, 2.0, 6.0],
    "Total": [None, None, None]
})

arquivo = "relatorio_tabela.xlsx"

# 1. Salva DataFrame
df.to_excel(arquivo, index=False, sheet_name="Dados")

# 2. Abre com openpyxl
wb = load_workbook(arquivo)
ws = wb["Dados"]

# 3. Insere fórmula NORMAL linha a linha
for row in range(2, ws.max_row + 1):
    ws[f"D{row}"] = f"=B{row}*C{row}"

# 4. Cria a tabela
tabela = Table(
    displayName="TabelaDados",
    ref=f"A1:D{ws.max_row}"
)

tabela.tableStyleInfo = TableStyleInfo(
    name="TableStyleMedium9",
    showFirstColumn=False,
    showLastColumn=False,
    showRowStripes=True,
    showColumnStripes=False
)

ws.add_table(tabela)

# 5. Salva
wb.save(arquivo)

print("Tabela criada — fórmula replica corretamente 🎉")
```

# Relatório com gráfico automatico direto no excel 

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.chart import BarChart, Reference

# --- depois de já ter criado a planilha e a tabela ---

arquivo = "relatorio_tabela.xlsx"

wb = load_workbook(arquivo)
ws = wb["Dados"]

# Criar gráfico
grafico = BarChart()
grafico.title = "Total por Produto"
grafico.y_axis.title = "Valor"
grafico.x_axis.title = "Produto"

# Dados (Total)
dados = Reference(
    ws,
    min_col=4,  # coluna Total
    min_row=1,
    max_row=ws.max_row
)

# Categorias (Produto)
categorias = Reference(
    ws,
    min_col=1,
    min_row=2,
    max_row=ws.max_row
)

grafico.add_data(dados, titles_from_data=True)
grafico.set_categories(categorias)

# Posicionar gráfico
ws.add_chart(grafico, "F2")

wb.save(arquivo)

print("Gráfico criado com sucesso!")
```
